```python
from PIL import Image
import numpy as np
import torch
import torch.nn.functional as F
```

```python
A_path = '../checkpoints/cifar10_resnet_baseline/adv_images.pth'
B_path = '../checkpoints/cifar10_resnet_ciiv/adv_images.pth'
```

```python
def show_image(data, idx, itype='org'):
    if itype == 'org':
        images = data['org_images']
    elif itype == 'adv':
        images = data['adv_images']
    img = (images[idx] * 255).long().permute(1,2,0).numpy().astype('uint8')
    return Image.fromarray(img, 'RGB').resize((256,256))
```

```python
def show_adv(data, idx):
    images = ((data['org_images'] - data['adv_images']) + 8) * 16
    img = (images[idx] * 255).long().permute(1,2,0).numpy().astype('uint8')
    return Image.fromarray(img, 'RGB').resize((256,256))
```

```python
A_data = torch.load(A_path)
B_data = torch.load(B_path)
```

```python
idx = 1
```

```python
show_image(B_data, idx, itype='org')
```

```python
show_adv(A_data, idx)
```

```python
show_adv(B_data, idx)
```
